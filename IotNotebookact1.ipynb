```python
import numpy as np
import matplotlib.pyplot as plt

##Generar senal en tiempo con rango desde -25 hasta 25 con un incremento de 0.01
t = np.arange(-25,25,0.01) 

#Funcion para generar la senal continua
xt = (2.5*np.cos(0.1*np.pi*t))+2.5
#xt = 130*np.cos(0.1*np.pi*t)

#Grafica desde el -25 hasta 25 en X y -130 hasta 130 en Y
plt.axis([-25,25,-10,10])
plt.ylabel("Y Anplitud")  
plt.xlabel("X Tiempo")
plt.plot(t,xt)
plt.grid()
plt.show()
```

```python
n = np.arange(-25,25,1)
xn = (127*np.cos(0.1*np.pi*n)/50)+2.5
plt.axis([-25,25,-10,10])
plt.ylabel("Y Volts")
plt.xlabel("X Nmuestras")
plt.stem(n,xn)
plt.grid()
plt.show()
```
